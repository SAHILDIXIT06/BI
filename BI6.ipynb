```python
import os 
os.environ["OMP_NUM_THREADS"] = "1"

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. Generate a professional dataset for clustering
# We create 300 samples with 4 distinct centers (clusters)
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# 2. Initialize and Apply K-Means
# We choose n_clusters=4 because we generated 4 centers
kmeans = KMeans(n_clusters=4, random_state=0, n_init='auto')
kmeans.fit(X)

# 3. Get results
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# 4. Visualization
plt.figure(figsize=(8, 6))

# Plot the data points colored by their cluster label
plt.scatter(X[:, 0], X[:, 1], c=labels, s=30, cmap='viridis', alpha=0.7)

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')

# Formatting the output
plt.title("K-Means Clustering Performance")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Show the results
plt.show()

# Print summary
print(f"Cluster Labels for first 10 points: {labels[:10]}")
print(f"Final Centroids Location:\n {centroids}")
```
